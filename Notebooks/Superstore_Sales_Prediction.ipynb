```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperstore.csv', encoding='ISO-8859-1')
```

```python
df.head()
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
```

```python
sns.barplot(x='Region',y='Sales',data=df)
```

```python

region_sales = df.groupby('Region')['Sales'].sum().reset_index()

sns.barplot(data=region_sales, x='Region', y='Sales', palette='viridis')
```

```python
df.info()
```

```python
df.drop(['Row ID','Order ID','Customer ID','Customer Name','Product Name'],axis=1,inplace=True)
```

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

df = df[(df['Sales']>lower_bound)&(df['Sales']<upper_bound)]
```

```python
df.info()
```

```python
df.drop(['Ship Date'],axis=1,inplace=True)
```

```python
df['Order Year'] = df['Order Date'].dt.year
```

```python
df.info()
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Day'] = df['Order Date'].dt.day
df.drop(['Order Date'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df_encoded = pd.get_dummies(df,drop_first=True)
```

```python
df_encoded.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(df_encoded.drop('Sales',axis=1),df_encoded['Sales'],test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42)
}

results={}
predictions={}

for name,model in models.items():
    model.fit(X_train,Y_train)
    y_pred = model.predict(X_test)
    predictions[name]=y_pred
    results[name]={
        "MAE": mean_absolute_error(Y_test, y_pred),
        "RMSE": np.sqrt(mean_squared_error(Y_test, y_pred)),
        "RÂ²": r2_score(Y_test, y_pred)
    }

df_results = pd.DataFrame(results).T

print(df_results)

```

```python
plots = plt.subplot()

for name, y_pred in predictions.items():
    plt.figure(figsize=(3, 2))
    sns.scatterplot(x=Y_test, y=y_pred)
    plt.xlabel("Actual Sales")
    plt.ylabel("Predicted Sales")
    plt.title(f"Actual vs Predicted - {name}")
    plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--')  # Perfect fit line
    plt.show()
```

```python
metrics_df = pd.DataFrame(results).T.reset_index().rename(columns={'index':'Model'})

plt.figure(figsize=(8,5))
sns.barplot(x='Model', y='RÂ²', data=metrics_df, palette='viridis')
plt.title("RÂ² Score Comparison Across Models")
plt.ylim(0, 1)
plt.show()
```

```python
df_results = pd.DataFrame({
    "Actual Sales": Y_test,
    "Linear Regression": predictions['Linear Regression'],
    "Decision Tree": predictions['Decision Tree'],
    "Random Forest": predictions['Random Forest'],
    "Gradient Boosting": predictions['Gradient Boosting'] 
})
df_results.head()
```

```python
df_results.to_csv("SuperStore_Predictions.csv",index=False)
```
